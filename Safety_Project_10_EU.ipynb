```python
import pandas as pd

# Filter the DataFrame for South West Europe
south_west_europe_data = merged_df_eu_wh_prod_jobch[
    merged_df_eu_wh_prod_jobch['Country Group'] == 'South West Europe'
]

# Step 1: Count incidents by Location and Week for South West Europe
incidents_per_week = south_west_europe_data.groupby(['Location', 'Week']).size().reset_index(name='Incident Count')

# Step 2: Calculate total incidents for South West Europe
total_incidents = incidents_per_week['Incident Count'].sum()

# Step 3: Calculate incident likelihood probability
incidents_per_week['Incident Probability'] = incidents_per_week['Incident Count'] / total_incidents

# Normalize the probabilities using the max probability for scaling
max_probability = incidents_per_week['Incident Probability'].max()
incidents_per_week['Normalized Probability'] = incidents_per_week['Incident Probability'] / max_probability

# Define danger levels based on normalized probability
def classify_danger(level):
    if level >= 0.75:
        return 'High'
    elif level >= 0.5:
        return 'Medium'
    else:
        return 'Low'

# Apply the classification
incidents_per_week['Danger Level'] = incidents_per_week['Normalized Probability'].apply(classify_danger)

# Sort the results by 'Normalized Probability' in descending order
incidents_per_week_sorted = incidents_per_week.sort_values(by='Normalized Probability', ascending=False)

# Display the top 3 weeks with the highest normalized probabilities
top_3_weeks = incidents_per_week_sorted.head(3)
print("Top 3 Dangerous Weeks in South West Europe:")
top_3_weeks
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates

# Ensure 'Year' and 'Month' are integers before constructing datetime objects
def ensure_integers(df, columns):
    for column in columns:
        df[column] = pd.to_numeric(df[column], downcast='integer')
    return df

# Function to plot data for a given country group
def plot_country_group(data, country_group):
    # Filter data for the given country group from 2021 to October 2024
    data_cg = data[
        (data['Country Group'] == country_group) &
        (data['Year'] >= 2021) &
        ((data['Year'] < 2024) | ((data['Year'] == 2024) & (data['Month'] <= 10)))
    ]

    # Count incidents per month by counting rows
    monthly_incidents = data_cg.groupby(['Year', 'Month']).size().reset_index(name='Incident Count')

    # Merge monthly incidents with tonnes packed to calculate rates
    monthly_data = monthly_incidents.merge(
        data_cg.groupby(['Year', 'Month'])['Tonnes Packed'].sum().reset_index(),
        on=['Year', 'Month'],
        how='left'
    )

    # Ensure 'Year' and 'Month' are treated as integers
    monthly_data = ensure_integers(monthly_data, ['Year', 'Month'])

    # Calculate incident rates: incidents / tonnes packed
    monthly_data['Incident Rate'] = monthly_data['Incident Count'] / monthly_data['Tonnes Packed']

    # Convert Year and Month to a single datetime column 'Date' for better plotting
    monthly_data['Date'] = pd.to_datetime(monthly_data['Year'].astype(str) + '-' + monthly_data['Month'].astype(str) + '-1', format='%Y-%m-%d')

    # Sort the data by 'Date'
    monthly_data.sort_values('Date', inplace=True)

    # Plotting
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plotting the monthly incident rates
    ax.plot(monthly_data['Date'], monthly_data['Incident Rate'], label='Monthly Incident Rate', marker='o', linestyle='-', color='blue')

    # Calculate and plot yearly mean lines
    years = monthly_data['Year'].unique()
    for year in years:
        year_data = monthly_data[monthly_data['Year'] == year]
        if not year_data.empty:
            mean_rate = year_data['Incident Rate'].mean()
            start_date = year_data['Date'].min()
            end_date = year_data['Date'].max()
            ax.plot([start_date, end_date], [mean_rate, mean_rate], 'r--', linewidth=1.5, label=f'Mean {year}')

    # Format the x-axis to show months and years
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

    # Setting labels and title
    ax.set_xlabel('Date')
    ax.set_ylabel('Incident Rate')
    ax.set_title(f'Monthly Incident Rates with Yearly Means for {country_group}')

    # Rotate date labels for better readability
    plt.xticks(rotation=45)
    plt.grid(True)

    # Improve the legend to avoid duplicate labels
    handles, labels = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels, handles))
    # ax.legend(by_label.values(), by_label.keys())

    plt.show()

# Loop over each country group and generate a plot
for country_group in merged_df_eu_wh_prod_jobch['Country Group'].unique():
    plot_country_group(merged_df_eu_wh_prod_jobch, country_group)
```

```python
import pandas as pd

# Filter for the year 2024 and specific country groups
filtered_df = merged_df_eu_wh_prod_jobch[
    (merged_df_eu_wh_prod_jobch['Year'] == 2024) &
    (merged_df_eu_wh_prod_jobch['Country Group'].isin([
        'South West Europe',
        'North Central Europe',
        'South East Europe',
        'United Kingdom Europe'])
    )
]

# Group by 'Country Group' and 'Location', then count rows
location_counts = filtered_df.groupby(['Country Group', 'Location']).size().reset_index(name='Incidents')

# Pivot the data to have locations as columns and counts as values
pivot_table = location_counts.pivot(index='Country Group', columns='Location', values='Incidents').fillna(0)

# Rename columns to reflect the specific incidents per location
pivot_table.columns = [f"{col} Incidents" for col in pivot_table.columns]

pivot_table
```
